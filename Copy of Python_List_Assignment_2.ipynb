### Question 1: Write a Python program to count the number of elements in a list without using len.


- `sample input : `
    - `var = [10,11,12,13]`
- `sample output :`
    - `result : 4`

```python
# Please write The solution Here for above question 1:

a = [10, 20, 30, 40, 50]
count = 0
for i in a:
  count += 1

print(f'result : {count}')
```

### Question 4: Write a Python program to find the average of all the elements in a list.


- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 2.833 `

```python
# Please write The solution Here for above question 4:

a = [1,2,3,4,2,5]
sum = 0

for i in a:
  sum += i

avg = sum/len(a)
print(f'Average -> {avg:.3f}')
```

### Question 6:  Write a Python program to find the index of a specified element in a list..

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 3 index is 2`

```python
# Please write The solution Here for above question 6:

a = [1,2,3,4,5,6,7,8,9]

x = 5

for i in range(len(a)):
  if a[i] == x:
    print(f'result : {x} index is {i}')
```

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
# Please write The solution Here for above question 7:

a = [121, 134, 155, 262]

for i in a:
  if str(i) == str(i)[::-1]:
    print(f'{i} is a palindrome')
  else:
    print(f'{i} is not a palindrome')
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
# Please write The solution Here for above question 8:

a = [1,2,3,4,10,11]
b = [10,11,12,14]
res = []
for i in a:
  if i in b:
    res.append(i)
for i in res:
  print(i)
```

### Question 9: Write a Python program to remove the duplicates from a list

- `sample input : `
    - `var = [10,11,12,10,12,24]`
- `sample output :`
    - `result : [10,11,12,24]`

```python
# Please write The solution Here for above question 9:

var = [10, 11, 12, 10, 12, 24]
res = []
for i in var:
  if i not in res:
    res.append(i)

print(f'result : {res}')
```
